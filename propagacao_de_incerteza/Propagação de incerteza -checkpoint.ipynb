O Tracker nos da uma medida de tempo com menor escala em milésimo de segundo, portanto, a incerteza associada ao tempo é de 0.001 s.
Já para a distância percorrida temos "$stotal$". As gotas que percorrem o percurso mais de uma vez terão seus valores de percurso na forma $stotal\times n$ sendo $n=2,3,4$.

Todas os cálculos de propagação de incerteza deste trabalho seguirão a seguinte metodologia:

$$\sigma_{t}=\sqrt{\left(\frac{\partial t}{\partial x_{1}}\sigma_{x_{1}}\right)^{2}+\left(\frac{\partial t}{\partial x_{2}}\sigma_{x_{2}}\right)^{2}+ ... + \left(\frac{\partial t}{\partial x_{n}}\sigma_{x_{n}}\right)^{2}}$$

Na qual $x_{i}$ com $i=1, 2, ... , n$ são as variáveis com suas respectivas incertezas e $t=t(x_{i})$.

### Cálculo dos valores da velocidade

```python
vs = y/ts #Velocidade de subida das gotas

vd = y/td #Velocidade de descida da gotas
```

```python
#Fazendo os valores das velocidades das gotas normalizados
vsnorm = unumpy.nominal_values(vs)*10**(4)
deltavsnorm = unumpy.std_devs(vs)*10**(4)
vdnorm = unumpy.nominal_values(vd)*10**(4)
deltavdnorm = unumpy.std_devs(vd)*10**(4)
```

### Cálculo do raio da gota

```python
#Adicionando os valores dados no roteiro experimental
po = 1030 #Densidade do óleo de silicone
pa = 1.293 #Densidade do ar a temperatura ambiente
n = 0.0000182 #Coeficiente de viscosidade do ar
g = 9.79 #Aceleração da gravidade local
d = ufloat(0.00250,0.00001) #Distância entre as placas do capacitor com incerteza
```

Dado que alguns valores utilizados são constantes, podemos compilar todos em uma constante só e posteriormente calcular o raio, logo

```python
# Constante de proporçao no cálculo do raio
k =(3/2)*((n/(g*(po-pa)))**(1/2))
k 
```

O módulo Uncertainties calcula a propagação de erro através do método de derivadas parciais descrito acima, portanto, podemos utilizá-lo a fim de diminuir o risco de cálculos errôneos.

```python
#Cálculo dos raios das gotas
r = k*(abs(vs-vd))**(1/2) 
```

```python
#Fazendo os valores dos raios das gotas normalizados
raio = unumpy.nominal_values(r)*10**7
deltar = unumpy.std_devs(r)*10**7
```

### Cálculo da carga das gotas

```python
q = ((9*np.pi*d)/2)*(((n**3)/(g*(po-pa)))**(1/2))*(((vs+vd)/uv)*((abs(vs-vd))**(1/2)))
```

```python
#Fazendo os valores da carga da gota normalizados
carga = unumpy.nominal_values(q)*10**(19)
deltaq = unumpy.std_devs(q)*10**(19)
```
